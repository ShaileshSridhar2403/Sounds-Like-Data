```python
#Importing required datasets.
import math
import pandas as pd
import sklearn.linear_model as sk
from sklearn.metrics import classification_report 
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

#Functions.
def weightCalc(docList):
    N = len(docList)
    df =[0 for i in docList[0]]
    for doc in docList:
        for i in range(len(docList[0])):
            df[i]+= int(doc[i]>0)
    df= [math.log(N/df[i]) if df[i]!=0 else 0  for i in range(len(df))]
    return df
    
def tf_idf(document,df):
    return np.multiply(document,df)

#Importing dataset.
fp = open("../Data/mxm_dataset_train.txt","r", encoding = 'utf-8');
fp2 = open("../Data/mxm_779k_matches.txt","r", encoding = 'utf-8');
BBdf = pd.read_csv("../Data/HotSongsBillBoard.csv", encoding = 'utf-8')

s = fp.readline();
s = s[1:]
lWords = s.split(',');
lWords.insert(0,"NULL")

d = {}
songList = []
songVectorList =[]
tmp = fp.readline()
while tmp:
    songList.append(tmp)
    tmp = fp.readline()
counter = 0
for i in range(19):
    tmp = fp2.readline()
while(tmp):
    l = tmp.split("<SEP>")
    d[l[0]] = l[2]
    tmp = fp2.readline()
    
    
    
for i in songList[:10000]:
    print("Row",counter,"done.")
    counter+=1
    countList = i.split(",")
    vector = [0 for i in range(len(lWords))]
    for j in countList[2:]:
        vals = j.split(":")
        index = int(vals[0])
        count = int(vals[1])
        vector[index] = count;
    songVectorList.append((countList[0],countList[1],vector))
        
df = pd.read_csv("../Data/Million_Song_Subset.csv")
```

```python
###########Creating vector list with hotttness as popularity metric
rows = []
y =[]
finList = []
for i in songVectorList:
  if i[0] in d.keys():
      row = df.loc[df['title'] == d[i[0]]]
      if row.shape[0] ==1:
          rows.append(row)
          finList.append((i,row['artist.hotttnesss'].iloc[0]));
          print(row)
```

```python
#######Test,train split for normal vectors, training with logistic regression
print("Carrying out logistic regression with hotttness")
x = [i[0][2] for i in finList]
y = [int(i[1]>0.4) for i in finList ]
xtrain = x[:int(0.85*len(x))]
xtest = x[int(0.85*len(x)):]
ytrain = y[:int(0.85*len(x))]
ytest = y[int(0.85*len(x)):]
clf = sk.LogisticRegression(random_state=0, solver='lbfgs').fit(xtrain, ytrain)
print(clf.score(xtest,ytest))
print(classification_report(ytest,clf.predict(xtest)))
```

```python
#######Test,train split for tf idf vectors, training with logistic regression
print("Carrying out logistic regression with hotttness(tf-idf)")
dF = weightCalc(x)
tf_idf_x = [tf_idf(i,dF) for i in x]
tf_idf_xtrain = tf_idf_x[:int(0.85*len(x))]
tf_idf_xtest = tf_idf_x[int(0.85*len(x)):]
clf_tf_idf = sk.LogisticRegression(random_state=0, solver='lbfgs').fit(xtrain, ytrain)
print(clf_tf_idf.score(tf_idf_xtest,ytest))
print(classification_report(ytest,clf_tf_idf.predict(tf_idf_xtest)))
```

```python
#######Test,train split for normal vectors, training with KNN
print("Carrying out knn with hotttness")
Knn= KNeighborsClassifier(n_neighbors=30)
Knn.fit(xtrain,ytrain)
print(Knn.score(xtest,ytest))
print(classification_report(ytest,Knn.predict(xtest)))
```

```python
#######Test,train split for tf-idf vectors, training with KNN
print("Carrying out knn with hotttness(tf-idf)")

Knn.fit(tf_idf_xtrain,ytrain)
print(Knn.score(tf_idf_xtest,ytest))
print(classification_report(ytest,Knn.predict(tf_idf_xtest)))
```

```python
#######Test,train split for normal vectors, training with Random Forest
print("Carrying out random forest with hotttness")
rf = RandomForestClassifier(n_estimators=1000, max_depth=100,random_state=0)
rf.fit(xtrain,ytrain)
print(rf.score(xtest,ytest))
print(classification_report(ytest,rf.predict(xtest)))
```

```python
#######Test,train split for tf idf vectors, training with Random Forest
print("Carrying out random forest with hotttness(tf-idf)")
rf = RandomForestClassifier(n_estimators=1000, max_depth=100,random_state=0)
rf.fit(tf_idf_xtrain,ytrain)
print(rf.score(tf_idf_xtest,ytest))
print(classification_report(ytest,rf.predict(tf_idf_xtest)))
```
